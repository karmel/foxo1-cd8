'''
There are many genes of interest with regard to the WT vs. KO phenotype.

We compare here.
'''

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
filename = '../data/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0)
data.columns
```

```python
# Clean up column names for easier parsing.
new_index = []
for col in data.columns:
    cleaned = col.replace('-KAA-14-10-31 reads','').replace(' FPKM','')
    cleaned = cleaned.replace('CD8TCell-','').replace('RNA-','')
    new_index.append(cleaned)
new_index[0] = 'id'
data.columns = new_index
data.columns
```

```python
# Extract the genes we care about.
factor_pairs = ((('Tcf7'),('NM_009331')),
           (('Eomes'),('NM_010136')),
           (('Tbx21'),('NM_019507')),
           (('Bcl2'),('NM_009741')),
           (('Id2'),('NM_010496')),
           (('Id3'),('NM_008321')),
           (('Prdm1'),('NM_007548')),
           (('Ccl5'),('NM_013653')),
           (('Klrg1'),('NM_016970')),
           (('Bcl6'),('NM_009744')),
           (('Stat3'),('NM_011486')),
           (('Stat4'),('NM_011487')),
           (('Ccr7'),('NM_030693')),
           (('Il7r'),('NM_008372')),
           (('Sell'),('NM_011346')),
           (('Nur77'),('NM_010444')),
           (('Nr4a2'),('NM_013613')),
           (('Klf4'),('NM_010637')),
           (('Coro2a'),('NM_001164804')),
           (('Ldlr'),('NM_010700')),
           (('Ly6c1'),('NM_001252055')),
           (('Ly6c2'),('NM_001099217')),
           (('Il18rap'),('NM_010553')),
           (('Ctla4'),('NM_001281976')),
           (('Cd55'),('NM_010016')),
           (('Rps3a1'),('NM_016959')),
           (('Icos'),('NM_017480')),
           (('Sema7a'),('NM_011352')),
           (('Gzmb'),('NM_013542 ')),
           
           )
zipped = list(zip(*factor_pairs))
factors = pd.DataFrame({'gene_name':zipped[0],
                        'id':zipped[1],})
factors
```

```python
subdata = pd.merge(factors, data, on='id', how='inner')
subdata.index = subdata['gene_name']
subdata
```

```python
# Now split to overlay WT and KO
wt = subdata[['LCMV-0d', 'KLRGhi-LCMV-7d', 'KLRGlo-LCMV-7d', 
        'KLRGhi-LCMV-12d', 'KLRGlo-LCMV-12d', 'LCMV-35d-rep2']]
ko = subdata[['Foxo1-KO-LCMV-0d', 'Foxo1-KO-KLRGhi-LCMV-7d', 
        'Foxo1-KO-KLRGlo-LCMV-7d', 'Foxo1-KO-KLRGhi-LCMV-12d']]
ko.index = ['Foxo1-KO-' + g for g in ko.index]

sns.set_palette('hls',2)

for gene in wt.index:
    plt.figure()
    ko.loc['Foxo1-KO-' + gene].plot()
    wt.loc[gene].plot()
    plt.title(gene + ': WT vs. KO')
    plt.ylabel('RPKM')
    plt.legend()
    
    plt.xticks(rotation=90)
    
    plt.savefig('figures/KO_vs_WT_{}.png'.format(gene),
            bbox_inches='tight')
```

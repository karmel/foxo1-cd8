'''
We have three variables of interest-- timecourse, KLRGhi vs. KLRGlo,
and Foxo1 KO vs. WT. Which generates the most variation in the 
dataset?

Use PCA to evaluate the axes of variation.
'''

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
sns.set_context('poster')
```

```python
filename = '../data/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0)
# Clean up column names for easier parsing.
new_index = []
for col in data.columns:
    cleaned = col.replace('-KAA-14-10-31 reads','').replace(' FPKM','')
    cleaned = cleaned.replace('CD8TCell-','').replace('RNA-','')
    new_index.append(cleaned)
new_index[0] = 'id'
data.columns = new_index
cols = ['LCMV-0d', 
        'KLRGhi-LCMV-7d', 
        'KLRGlo-LCMV-7d', 
        'KLRGhi-LCMV-12d', 
        'KLRGlo-LCMV-12d', 
        'LCMV-35d-rep2', 
        'Foxo1-KO-LCMV-0d',
        'Foxo1-KO-KLRGhi-LCMV-7d', 
        'Foxo1-KO-KLRGlo-LCMV-7d', 
        'Foxo1-KO-KLRGhi-LCMV-12d', 
        ]
subdata = data[cols]
subdata.index = data['id']
subdata.head()
```

```python
# Filter out genes that aren't changing.
def check_changing(row):
    if row.max() - row.min() < row.min():
        return False
    if row.max() < 10:
        return False
    return True
filtered = subdata[subdata.apply(check_changing, axis=1)]
filtered.head()
```

```python
pca = PCA(n_components=3)
X_r = pca.fit(filtered).transform(filtered)
print('Explained variance ratio (first three components): '
      + str(pca.explained_variance_ratio_))
X_r
```

```python
plt.figure()
plt.scatter(X_r[:, 0], X_r[:, 1])
plt.title('PCA')
```

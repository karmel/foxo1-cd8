'''
What sorts of gene clusters fall out of the data?

Lots of different subsets of columns to try. 
Let's start with Affinity Propagation.
'''

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
# Set up some rules.
max_fdr = 0.01
min_logfc = 1
min_logcpm = 2
```

```python
filename = '../data/diffExp.output.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0)
data.index = data[data.columns[0]].tolist()

# Clean up column names for easier parsing.
cols = ['KLRGhi-LCMV-7d vs. RNA-LCMV-0d ',
        'KLRGlo-LCMV-7d vs. RNA-LCMV-0d ',
        'KLRGhi-LCMV-12d vs. RNA-LCMV-0d ',
        'KLRGlo-LCMV-12d vs. RNA-LCMV-0d ',
        'RNA-LCMV-0d vs. RNA-LCMV-35d-rep2 ',
        'RNA-LCMV-0d vs. Foxo1-KO-LCMV-0d ',
        'RNA-LCMV-0d vs. Foxo1-KO-KLRGhi-LCMV-7d ',
        'RNA-LCMV-0d vs. Foxo1-KO-KLRGlo-LCMV-7d ',
        'RNA-LCMV-0d vs. Foxo1-KO-KLRGhi-LCMV-12d ',

]
# Get three data points for each prefix, then flatten
all_cols = [(pre + 'logFC', pre + 'logCPM', pre + 'FDR') for pre in cols]
all_cols = [item for sublist in all_cols for item in sublist]
subdata = data[all_cols]
subdata.describe()

```

```python
# Filter out genes that aren't changing.
mask_fdr = subdata.filter(like='FDR').min(axis=1) <= max_fdr
mask_cpm = subdata.filter(like='logCPM').max(axis=1) >= min_logcpm
mask_fc = subdata.filter(like='logFC').max(axis=1).abs() >= min_logfc
filtered = subdata[mask_fdr & mask_cpm & mask_fc]
filtered.describe()
```

```python
# Isolate logFC values
logfc = filtered.filter(like='logFC')
logfc.head()
```

```python
# Flip the sign so that all values are relative to RNA-LCMV-0d
for col in logfc.columns:
    if 'vs. RNA-LCMV-0d' in col:
        logfc[col] = - logfc[col]
logfc.head()
```

```python
scaled = scale(logfc)
```

```python
models = {}
min_clusters = None
min_cluster_af = None
for pref in (-500, -400, -300, -200, -100, -75, None):
    af = AffinityPropagation(preference=pref).fit(scaled)
    n_clusters_ = len(af.cluster_centers_indices_)

    if not min_clusters or min_clusters > n_clusters_:
        min_clusters = n_clusters_
        min_cluster_af = af
    print('Estimated number of clusters for pref={}: {}'.format(pref, n_clusters_))
    models[pref] = af
labels = min_cluster_af.labels_
```

```python
# Make dataframe with our scaled data.
labels = models[-300].labels_
scaled_df = pd.DataFrame(scaled, index=logfc.index, columns=logfc.columns)
scaled_df.head()
```

```python
# Add in gene names and labels for easier viewing
descriptions = data[mask_fdr & mask_cpm & mask_fc]['Annotation/Divergence']
gene_names = [s.split('|')[0] for s in descriptions]
scaled_df['gene_name'] = gene_names
scaled_df['label'] = labels
scaled_df.head()
```

```python
# Output sorted by cluster.
clustered = scaled_df.sort(columns=['label', 'KLRGhi-LCMV-7d vs. RNA-LCMV-0d logFC'])
clustered.head()
```

```python
clustered.to_csv('output/affinity_clusters.txt', sep='\t',
                 header=True, index=True,
                 columns=['gene_name', 'label'] + list(clustered.columns[:-2]))
```

'''
What sorts of gene clusters fall out of the data?

Lots of different subsets of columns to try. 
Let's start with Affinity Propagation.
'''

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
filename = '../data/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0)
# Clean up column names for easier parsing.
new_index = []
for col in data.columns:
    cleaned = col.replace('-KAA-14-10-31 reads','').replace(' FPKM','')
    cleaned = cleaned.replace('CD8TCell-','').replace('RNA-','')
    new_index.append(cleaned)
new_index[0] = 'id'
data.columns = new_index
data.index = data['id']
cols = ['LCMV-0d', 
        'KLRGhi-LCMV-7d', 
        'KLRGlo-LCMV-7d', 
        'KLRGhi-LCMV-12d', 
        'KLRGlo-LCMV-12d', 
        'LCMV-35d-rep2', 
        'Foxo1-KO-LCMV-0d',
        'Foxo1-KO-KLRGhi-LCMV-7d', 
        'Foxo1-KO-KLRGlo-LCMV-7d', 
        'Foxo1-KO-KLRGhi-LCMV-12d', 
        ]
subdata = data[cols]
subdata.head()
```

```python
# Filter out genes that aren't changing.
def check_changing(row):
    if row.max() - row.min() < row.min():
        return False
    if row.max() < 10:
        return False
    return True
mask = subdata.apply(check_changing, axis=1)
filtered = subdata[mask]
filtered.head()
```

```python
scaled = scale(filtered)
```

```python
af = AffinityPropagation(preference=-50).fit(scaled)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)

print('Estimated number of clusters: %d' % n_clusters_)
labels
```

```python
# Make dataframe with our scaled data.
scaled_df = pd.DataFrame(scaled, index=filtered.index, columns=filtered.columns)
scaled_df.head()
```

```python
# Add in gene names and labels for easier viewing
gene_names = [s.split('|')[0] for s in data[mask]['Annotation/Divergence']]
scaled_df['gene_name'] = gene_names
scaled_df['label'] = labels
scaled_df.head()
```

```python
# Output sorted by cluster.
clustered = scaled_df.sort(columns=['label', 'LCMV-0d'])
clustered.head()
```

```python
clustered.to_csv('output/affinity_clusters.txt', sep='\t',
                 header=True, index=True,
                 columns=['gene_name', 'label'] + list(clustered.columns[:-2]))
```
